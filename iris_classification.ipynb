```python
import torch
import  torch.nn as nn  
import numpy as np
from torch.optim import SGD, Adam
from torch.nn import Module , ReLU, Softmax, Sigmoid
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

sc = StandardScaler()
data = load_iris()

X , y = data.data, data.target

X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=42, test_size=0.2)

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

X_train = torch.from_numpy(X_train.astype(np.float32))
X_test = torch.from_numpy(X_test.astype(np.float32))
y_train = torch.from_numpy(y_train.astype(np.int64))
y_test = torch.from_numpy(y_test.astype(np.int64))

# print(X_train.shape)
# print(X_test.shape)
# print(y_train.shape)
# print(y_test.shape)

n_input_features = X_train.shape[1]

# Define the model architecture
class ComplexModel(nn.Module):
    def __init__(self, n_input_features):
        super(ComplexModel, self).__init__()
        self.fc1 = nn.Linear(n_input_features, 64)
        self.fc2 = nn.Linear(64, 3)
        
        # forward
    def forward(self, X):
        X = torch.relu(self.fc1(X))
        X = self.fc2(X)
        return torch.softmax(X,dim = 1)  

model = ComplexModel(n_input_features = n_input_features)

# hyperparameters 
epochs = 2000
learning_rate = 0.2
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr = learning_rate)

# training

for epoch in range(epochs):
    model.train()                       # set the model to training mode
    y_pred = model(X_train)
    loss = criterion(y_pred, y_train)


    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    if epoch%200== 0:
        print(f'epoch = {epoch+1:.2f}, loss = {loss.item():.4f}')

model.eval()       
with torch.no_grad():
    y_pred_test = model(X_test)
    loss = criterion(y_pred_test, y_test)
    print(f'Test loss {loss.item():.2f}')
```
